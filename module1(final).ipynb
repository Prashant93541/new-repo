### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
data = pd.read_csv("MindTree.csv")  
total=len(shares)
def avg_price(N):
    sum1=0
    for i in range(total-N,total):
        sum1+=data.iloc[i]['Close Price']     
    avg_price=sum1/N
    print(avg_price)
N=int(input())
avg_price(N)

def profitloss_percentage(N):
    total=len(shares)
    diff=0
    b=data.iloc[total-1]['Close Price']
    c=data.iloc[total-N]['Close Price']
    diff=b-c
    if(diff<0):
        loss_percent=(diff/b)*100
        print(abs(loss_percent)) 
    else:
        profit_percent=(diff/c)*100
        print(profit_percent)
N=int(input())
profitloss_percentage(N)

avg_price(7)
avg_price(14)
avg_price(30)
avg_price(60)
avg_price(180)
avg_price(365)
profitloss_percentage(7)
profitloss_percentage(14)
profitloss_percentage(30)
profitloss_percentage(60)
profitloss_percentage(180)
profitloss_percentage(365)
```

### Query1.6 
Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a ‘Nan’ value. Why does this happen? Either remove the first row, or set the entry to 0 before proceeding.

```python
shares['Day_Perc_Change']=(shares['Close Price'].pct_change()*100)
shares.Day_Perc_Change[0]=0 #this is required because the first value has no previous day's closing price data
shares
```

### Query 1.7 
Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
total=len(shares)
shares['Trend']=''
shares.loc[(shares['Day_Perc_Change']>-0.5)&(shares['Day_Perc_Change']<0.5),'Trend']='Slight or No Change'
shares.loc[(shares['Day_Perc_Change']>0.5)&(shares['Day_Perc_Change']<1),'Trend']='Slight positive'
shares.loc[(shares['Day_Perc_Change']>-1)&(shares['Day_Perc_Change']<-0.5),'Trend']='Slight Negative'
shares.loc[(shares['Day_Perc_Change']>1)&(shares['Day_Perc_Change']<3),'Trend']='Positive'
shares.loc[(shares['Day_Perc_Change']>-3)&(shares['Day_Perc_Change']<-1),'Trend']='Negative'
shares.loc[(shares['Day_Perc_Change']>3)&(shares['Day_Perc_Change']<7),'Trend']='Among top gainers'
shares.loc[(shares['Day_Perc_Change']>-7)&(shares['Day_Perc_Change']<-3),'Trend']='Among top losers'
shares.loc[shares['Day_Perc_Change']>7,'Trend']='Bull run'
shares.loc[shares['Day_Perc_Change']<-7,'Trend']='Bear drop'
shares['Trend']
```
